```python
import numpy as np
from PIL import Image


def grayscale_histogram(image: np.ndarray) -> np.ndarray:
    histogram = np.zeros(256, dtype=int)
    for pixel in image.flatten():
        histogram[pixel] += 1
    return histogram


def concat_rgb_histogram(image: np.ndarray) -> np.ndarray:
    hist = np.zeros((3, 256), dtype=int)

    for r, g, b in image.reshape(-1, 3):
        hist[0, r] += 1
        hist[1, g] += 1
        hist[2, b] += 1

    return hist.flatten()
```

```python
path = "./descriptor/BBDD/bbdd_00009.jpg"
image = np.array(Image.open(path))
grayscale_img = np.array(Image.open(path).convert("L"))

gray_out = grayscale_histogram(grayscale_img)
rgb_out = concat_rgb_histogram(image)

print("Grayscale Histogram:", gray_out, len(gray_out))
print("Concatenated RGB Histogram:", rgb_out, len(rgb_out))
```
